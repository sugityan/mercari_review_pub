```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import time
import tqdm
```

```python
# sort=new: 新しい順

url = "https://minhyo.jp/mercari?sort=new&page=1"
response = requests.get(url)
time.sleep(1)
soup = BeautifulSoup(response.text, "html.parser")

"""
口コミはインデックス1から13
内容は、contentsのインデックス1にある
"""
elems = soup.find_all("div", class_=re.compile("readmore-inner"))
elems[1].contents[1]

```

```python
df = pd.DataFrame()
cnt = 1
# pageが４までは、2023年度の口コミ
# pageが4から39までは2022年度の口コミ
for i in tqdm.tqdm(range(1, 5)):
    url = "https://minhyo.jp/mercari?sort=new&page={}".format(i)
    response = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(response.text, "html.parser")
    elems = soup.find_all("div", class_=re.compile("readmore-inner"))

    for j in range(1, 14):
        try:
            df = pd.concat([df, pd.DataFrame([elems[j].get_text()], index=[cnt])])
            cnt += 1
        except IndexError:
            continue
        except Exception as e:
            print("page number:", i)
            print("Error message:",e)
            print("Index Number:", j)
            df.to_csv("mercari_review23.csv")
            continue
        
df.to_csv("mercari_review_23.csv")
```

```python
df = pd.DataFrame()
cnt = 1
# pageが5から39までは2022年度の口コミ
for i in tqdm.tqdm(range(5, 39)):
    url = "https://minhyo.jp/mercari?sort=new&page={}".format(i)
    response = requests.get(url)
    time.sleep(1)
    soup = BeautifulSoup(response.text, "html.parser")
    elems = soup.find_all("div", class_=re.compile("readmore-inner"))

    for j in range(1, 14):
        try:
            df = pd.concat([df, pd.DataFrame([elems[j].get_text()], index=[cnt])])
            cnt += 1
        except IndexError:
            continue
        except Exception as e:
            print("page number:", i)
            print("Error message:",e)
            print("Index Number:", j)
            df.to_csv("mercari_review_22.csv")
            continue
df.to_csv("mercari_review_22.csv")
```
